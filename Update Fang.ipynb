```python

directory_in_str = 'fang'

directory = os.fsencode(directory_in_str)

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".csv"):
        print(filename)
        og = pd.read_csv(directory_in_str + '/' +filename, sep = ",")
        new = pd.read_csv('faang_up_to_date' + '/' +filename, sep = ",")
        og=og.drop(columns=['ticker','actual_diff','Unnamed: 0'])
        og = og.rename(str.lower, axis='columns')
        new = new.rename(str.lower, axis='columns')
        og.loc[:, 'date'] = pd.to_datetime(og['date'],format='%m/%d/%Y')
        new.loc[:, 'date'] = pd.to_datetime(new['date'],format='%Y-%m-%d')

        add = new[(new['date'] > '2018-08-24')]
        og =og.append(add)
        print(add.tail())
        print(og.tail())
        og.to_csv(directory_in_str + '/' +filename)
```
